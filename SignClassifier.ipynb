```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam, SGD, RMSprop
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import pandas as pd
import seaborn as sns
import kagglehub
from sklearn.preprocessing import LabelBinarizer
from sklearn.metrics import roc_curve, auc

# Step 1: Download dataset using kagglehub
path = kagglehub.dataset_download("meowmeowmeowmeowmeow/gtsrb-german-traffic-sign")
print("Dataset downloaded to:", path)

# Step 2: Define dataset paths
train_dir = os.path.join(path, 'Train')
test_csv_path = os.path.join(path, 'Test.csv')

# Step 3: Load and preprocess the training data
data = []
labels = []
classes = 43  # Total number of classes

# Loading training images by class
for i in range(classes):
    class_path = os.path.join(train_dir, str(i))  # Path to each class folder
    images = os.listdir(class_path)
    for img_file in images:
        try:
            img_path = os.path.join(class_path, img_file)
            image = Image.open(img_path)
            image = image.resize((30, 30))  # Resize to 30x30 pixels
            data.append(np.array(image))
            labels.append(i)
        except Exception as e:
            print(f"Error loading image {img_file}: {e}")

# Convert lists into numpy arrays
data = np.array(data)
labels = np.array(labels)

# Normalize the data using MinMaxScaler
scaler = MinMaxScaler()
ascolumns = data.reshape(-1, 3)  # Flatten images for normalization
t = scaler.fit_transform(ascolumns)
data = t.reshape(data.shape)  # Reshape back to the original shape

# Split data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(data, labels, test_size=0.2, random_state=42)

# One-hot encode the labels
y_train = to_categorical(y_train, classes)
y_val = to_categorical(y_val, classes)

# Function to plot accuracy vs. epochs
def plot_accuracy_vs_epochs(history, model_name):
    plt.plot(history.history['accuracy'], label=f'{model_name} training accuracy')
    plt.plot(history.history['val_accuracy'], label=f'{model_name} validation accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.title(f'{model_name} Accuracy vs. Epochs')
    plt.show()

# Confusion Matrix Heatmap
def plot_confusion_matrix(y_true, y_pred, classes, model_name):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title(f'Confusion Matrix - {model_name}')
    plt.show()

# ROC Curve (Receiver Operating Characteristic Curve)
def plot_roc_curve(y_true, y_pred_prob, classes, model_name):
    lb = LabelBinarizer()
    y_true_bin = lb.fit_transform(y_true)
    fpr, tpr, _ = roc_curve(y_true_bin.ravel(), y_pred_prob.ravel())
    roc_auc = auc(fpr, tpr)
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve - {model_name}')
    plt.legend(loc="lower right")
    plt.show()

# Step 4: Build CNN Model with Different Configurations
def build_model(learning_rate=0.001, optimizer='adam', filters=32, dropout_rates=[0.25, 0.25, 0.5], epochs=20, batch_size=64):
    model = Sequential([
        Conv2D(filters=filters, kernel_size=(5, 5), activation='relu', input_shape=X_train.shape[1:]),
        MaxPool2D(pool_size=(2, 2)),
        Dropout(rate=dropout_rates[0]),
        Conv2D(filters=filters*2, kernel_size=(3, 3), activation='relu'),
        MaxPool2D(pool_size=(2, 2)),
        Dropout(rate=dropout_rates[1]),
        Conv2D(filters=filters*4, kernel_size=(3, 3), activation='relu'),
        MaxPool2D(pool_size=(2, 2)),
        Dropout(rate=dropout_rates[2]),
        Flatten(),
        Dense(256, activation='relu'),
        Dropout(rate=0.5),
        Dense(classes, activation='softmax')
    ])
    if optimizer == 'adam':
        optimizer_instance = Adam(learning_rate=learning_rate)
    elif optimizer == 'sgd':
        optimizer_instance = SGD(learning_rate=learning_rate)
    else:
        optimizer_instance = RMSprop(learning_rate=learning_rate)
    model.compile(loss='categorical_crossentropy', optimizer=optimizer_instance, metrics=['accuracy'])
    return model

# Step 5: Train the model and evaluate on validation data
def train_and_evaluate(learning_rate=0.001, optimizer='adam', filters=32, dropout_rates=[0.25, 0.25, 0.5], epochs=20, batch_size=64):
    model = build_model(learning_rate, optimizer, filters, dropout_rates, epochs, batch_size)
    # Train the model
    history = model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_data=(X_val, y_val))
    # Plot accuracy vs epochs
    plot_accuracy_vs_epochs(history, f'Model {learning_rate} - {optimizer} - {epochs} epochs')
    # Get predictions and plot confusion matrix
    y_pred = np.argmax(model.predict(X_val), axis=1)
    plot_confusion_matrix(y_val.argmax(axis=1), y_pred, classes=[str(i) for i in range(43)], model_name=f'Model {learning_rate} - {optimizer}')
    # Get predicted probabilities for ROC curve
    y_pred_prob = model.predict(X_val)
    plot_roc_curve(y_val.argmax(axis=1), y_pred_prob, classes=[str(i) for i in range(43)], model_name=f'Model {learning_rate} - {optimizer}')
    return model, history

# Step 6: Evaluate model on the test set
def evaluate_test_model(model):
    # Load test data
    y_test = pd.read_csv(test_csv_path)
    y_test["ClassId"] = y_test["ClassId"].astype(int)
    test_labels = y_test["ClassId"].values
    test_imgs = y_test["Path"].values
    test_data_list = []
    for img_path in test_imgs:
        full_img_path = os.path.join(path, img_path)
        image = Image.open(full_img_path)
        image = image.resize((30, 30))
        test_data_list.append(np.array(image))
    X_test = np.array(test_data_list)
    # Normalize the test data
    ascolumns = X_test.reshape(-1, 3)
    t = scaler.transform(ascolumns)
    X_test = t.reshape(X_test.shape)
    # Predict and calculate evaluation metrics
    y_pred_test = np.argmax(model.predict(X_test), axis=1)
    accuracy = accuracy_score(test_labels, y_pred_test)
    precision = precision_score(test_labels, y_pred_test, average='macro')
    recall = recall_score(test_labels, y_pred_test, average='macro')
    f1 = f1_score(test_labels, y_pred_test, average='macro')
    print(f'Accuracy: {accuracy:.4f}')
    print(f'Precision: {precision:.4f}')
    print(f'Recall: {recall:.4f}')
    print(f'F1 Score: {f1:.4f}')
    return accuracy, precision, recall, f1

# Step 7: Experiment with Different Configurations
configs = [
    {'learning_rate': 0.001, 'optimizer': 'adam', 'filters': 32, 'dropout_rates': [0.25, 0.25, 0.5], 'epochs': 20, 'batch_size': 64},
    {'learning_rate': 0.0001, 'optimizer': 'adam', 'filters': 64, 'dropout_rates': [0.3, 0.3, 0.5], 'epochs': 50, 'batch_size': 32},
    {'learning_rate': 0.01, 'optimizer': 'sgd', 'filters': 32, 'dropout_rates': [0.3, 0.3, 0.6], 'epochs': 30, 'batch_size': 64},
    {'learning_rate': 0.001, 'optimizer': 'rmsprop', 'filters': 128, 'dropout_rates': [0.4, 0.4, 0.7], 'epochs': 25, 'batch_size': 64},
]

for config in configs:
    print(f"Training with configuration: {config}")
    model, history = train_and_evaluate(**config)
    print("\n")
    evaluate_test_model(model)
```
